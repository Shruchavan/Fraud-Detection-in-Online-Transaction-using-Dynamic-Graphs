```python
from neo4j import GraphDatabase
import pandas as pd

# Connect to Neo4j database
uri = "bolt://localhost:7687"
driver = GraphDatabase.driver(uri, auth=("neo4j", "Shru@2001"))

# Function to fetch transaction data with their relationships
def fetch_transaction_data(tx):
    query = """
    MATCH (t:Transaction)-[r:MADE|AT]->(n)
    RETURN t.trans_num AS trans_num, t.amt AS amount, t.is_fraud AS is_fraud, 
           t.timestamp AS timestamp, t.cc_num AS cc_num, 
           type(r) AS relationship_type, 
           n.name AS merchant_name, n.cc_num AS user_cc_num
    """
    result = tx.run(query)
    return pd.DataFrame([record.data() for record in result])

# Create a session and execute the query
with driver.session() as session:
    df = session.read_transaction(fetch_transaction_data)

# Display the extracted data
print(df.head())

# Close the driver connection
driver.close()
```

```python
from neo4j import GraphDatabase

def save_predictions_batch(tx, batch_df):
    for index, row in batch_df.iterrows():
        query = (
            "MATCH (t:Transaction {trans_num: $trans_num}) "
            "SET t.predicted_fraud = $predicted_fraud"
        )
        tx.run(query, trans_num=row['trans_num'], predicted_fraud=row['predicted_fraud'])

def process_in_batches(df, batch_size=1000):
    for start in range(0, len(df), batch_size):
        end = start + batch_size
        yield df.iloc[start:end]

# Connecting to Neo4j
driver = GraphDatabase.driver(uri="bolt://localhost:7687", auth=("neo4j", "Shru@2001"))

# Processing the transactions DataFrame in batches
batch_size = 1000  # Adjust batch size as necessary
with driver.session() as session:
    for batch_df in process_in_batches(transactions_df, batch_size):
        session.execute_write(save_predictions_batch, batch_df)

# Closing the driver
driver.close()
```

```python
from py2neo import Graph
import pandas as pd

# Connect to Neo4j database
graph = Graph("bolt://localhost:7687", auth=("neo4j", "Shru@2001"))

# Query for fraud transactions and related users
query = '''
MATCH (t:Transaction)-[:MADE]->(u:User)
WHERE t.predicted_fraud = 1
RETURN t.trans_num AS transaction_number, u.cc_num AS credit_card_number, t.amt AS amount
'''

# Execute the query and store results in 'data'
data = graph.run(query).to_data_frame()

```
